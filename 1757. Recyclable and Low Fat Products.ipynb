```python
import pandas as pd

# Input: Products table:
Products = {
    'product_id': ['0', '1', '2', '3', '4'],
    'low_fats':   ['Y', 'Y', 'N', 'Y', 'N'],
    'recyclable': ['N', 'Y', 'Y', 'Y', 'N']
}

# Create DataFrame 
df = pd.DataFrame(Products)

# Print the DataFrame
df
```

## **Question**

**Write a solution to find the ids of products that are both low fats and recyclable**

**Return the result in **any order****

```python
# Filter products that are both low fat and recyclable
filtered_df = df[(df['low_fats'] == 'Y') & (df['recyclable'] == 'Y')]

# Select only the 'product_id' column
filtered_df[['product_id']]
```
